###  Creating the Visualization

Finally we have to create the visualization, first let remove noise from the text data using the nltk stopwords, its imperative to do this, as this process removes pronoun and conjunction like `['me','we','my','and']`, we would also add the players names to the stopwords list, because it is expected that his name would appear frequently on the page, and this would lead to noise in the text data.

```python
stopwords_list = stopwords.words('english')
stopwords_list += list(string.punctuation)
stopwords_list += ['Messi', 'player', 'Lionel','ljoˈnel','anˈdɾes','pronunciation:','del','listen','Leo']
stopwords_list[:5]
```

To round it up we would be writing a function to implement our wordcloud function 

```python
def generate_better_wordcloud(data, title, mask=None):
    cloud = WordCloud(scale=3, max_words=150, colormap='Dark2',stopwords=stopwords_list, mask=mask,
                      background_color='white').generate(data)
    plt.figure(figsize=(10,8))
    plt.imshow(cloud, interpolation='bilinear')
    plt.axis('off')
    plt.title('\n'.join(wrap(title,60)), fontsize=18)
    plt.show()
```

```python
generate_better_wordcloud(sum_result,'Messi Wikipedia', mask = mask)
```
